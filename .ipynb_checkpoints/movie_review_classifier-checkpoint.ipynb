```python
documents = []
for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        documents.append((movie_reviews.words(fileid), category))

documents[0:5]
```

```python
# let's shuffle docs
import random
random.shuffle(documents)
documents[0:5]
```

```python
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet 
def get_simple_pos(tag):
    
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB 
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV 
    else :
        return wordnet.NOUN
```

```python
#nltk.download('averaged_perceptron_tagger')
pos_tag(['better'])
```

```python
pos_tag(['This','painting','looks','better'])
```

```python
# creating list of stop words 
#nltk.download('stopwords')
from nltk.corpus import stopwords 
import string 
stops = set(stopwords.words('english'))
punctuations = list(string.punctuation)
(punctuations)
stops.update(punctuations)
stops
```

```python
# cleaning function 

def clean_review(words) :
    output_words = []
    for w in words :
        if w.lower() not in stops : 
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w, pos = get_simple_pos(pos[0][1]))
            output_words.append(clean_word.lower())
    return output_words
```

```python
#nltk.download('wordnet')
documents = [ (clean_review(document), category) for document, category in documents ]
```

```python
(documents[1])
```

```python
len(documents)
```

```python
training_documents = documents[0:1500]
testing_documents = documents[1500:]
```

```python
all_words = []
for doc in training_documents:
    all_words += doc[0]
```

```python
import nltk 
```

```python
freq = nltk.FreqDist(all_words)
# freq : freq distribution object
type(freq) , freq
```

```python
len(freq), freq.most_common(15)
```

```python
common = freq.most_common(2500)
features = [item[0] for item in freq.most_common(2500)]
features
```

```python
def get_feature_dict(words) :
    current_features = {}
    word_set = set(words)
    for w in features :
        current_features[w] = w in word_set # important line of code for dictionary
        
    return current_features 
```

```python
output = get_feature_dict(training_documents[0][0])
output 
```

```python
# go through each document 
training_data = [(get_feature_dict(doc), category) for doc, category in training_documents ] 
len(training_data)
```

```python
testing_data = [(get_feature_dict(doc), category) for doc, category in testing_documents ]
len(testing_data)
```

```python
from nltk import NaiveBayesClassifier 
```

```python
classifier = NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(classifier, testing_data)
```

# Using count vectorizer on our documents data

```python
len(documents) , len(training_documents), len(testing_documents)
```

```python
documents[0][1] , training_documents[0][1], testing_documents[0][1]
```

```python
# refer appendix 3 for some tutorial 
```

```python
categories = [category for doc, category in documents]
```

```python
text_documents = [ " ".join(doc) for doc, category in documents ]
```

```python
(text_documents[0]) #list of documents  
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(text_documents, categories )
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vec = CountVectorizer(max_features = 2500, ngram_range = (1,2))
                           
x_train_features = count_vec.fit_transform(x_train)
x_train_features, x_train_features.todense()
```

```python
count_vec.get_feature_names()
```

```python
x_test_features = count_vec.transform(x_test)
x_test_features
```

```python
from sklearn.svm import SVC 
```

```python
svc = SVC(gamma="auto")
svc.fit(x_train_features,y_train)
```

```python
svc.score(x_test_features, y_test) 
# no ngram , and svm on 2500 gave 0.796
#(1,2) ngram , svm on 2500 data points gave 0.798
```
